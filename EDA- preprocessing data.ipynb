Now, we have to define the Bounding Box. Bounding Box is the area defined by two longitudes and two latitudes that will include all spatial points.

Choose data samples of any polygon that’s `at least` **10km x 10km** size (100 Km2).

```python
#calculate subset Polygon area
import pyproj    
import shapely
import shapely.ops as ops
from shapely.geometry.polygon import Polygon
from functools import partial


geom = Polygon([(25.0, 46.4),(24.85, 46.4), (24.85, 46.6), (25.0, 46.6)])
geom_area = ops.transform(
    partial(
        pyproj.transform,
        pyproj.Proj(init='EPSG:4326'),
        pyproj.Proj(
            proj='aea',
            lat_1=geom.bounds[1],
            lat_2=geom.bounds[3]
        )
    ),
    geom)

# Print the area in Km^2
print((geom_area.area)/1000000,'','Km2')
```

With only these 3 lines of code, Geometry column is created inside a Geopandas GeodataFrame.

```python
#1-RSRP
# creating a geometry column 
geometry = [Point(xy) for xy in zip(rsrp['LocationLongitude'], rsrp['LocationLatitude'])]
# Coordinate reference system : WGS84
crs = {'init': 'epsg:4326'}
# Creating a Geographic data frame 
gdf = gpd.GeoDataFrame(rsrp, crs=crs, geometry=geometry)
```

```python
#1-Traffic Volume
# creating a geometry column 
geometry = [Point(xy) for xy in zip(traf['LocationLongitude'], traf['LocationLatitude'])]
# Coordinate reference system : WGS84
crs = {'init': 'epsg:4326'}
# Creating a Geographic data frame 
gdf2 = gpd.GeoDataFrame(traf, crs=crs, geometry=geometry)
```

```python
from shapely.geometry import Polygon, Point

#adding a column that contains a boolean variable for each point
#RSRP data:
rsrp["polygon"] = rsrp.apply(lambda row: geom.contains(Point(row["LocationLongitude"], row["LocationLatitude"])), axis = 1)
#Traffic volume data
traf["polygon"] = traf.apply(lambda row: geom.contains(Point(row["LocationLongitude"], row["LocationLatitude"])), axis = 1)
print(rsrp.head())
```

## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly_express as px
import plotly.graph_objects as go
sns.set()
from pandas.io.json import json_normalize 
```

You can find out what categories exist and how many types belong to each category by using the value_counts() method.
we need to answer many questions like :

```python
#1-What is the most type of mobiles ?
rsrp["DeviceManufacturer"].value_counts()
```

```python
l= ['RadioOperatorName', 'DeviceManufacturer', 'DeviceName']


for i in l:
    la=rsrp[i].value_counts().index
    la=la[0:5]
    f, ax = plt.subplots(figsize=(7, 4))
    sns.countplot(x= i, data= rsrp, order= la )
    plt.figure()
```

```python
labels=rsrp['DeviceManufacturer'].value_counts().index
values=rsrp['DeviceManufacturer'].value_counts().values

#visualization
plt.figure(figsize=(7,7))
plt.pie(values ,labels = labels ,autopct='%1.1f%%')
plt.title('DeviceManufacturer')
plt.show()
plt.savefig('DeviceManufacturer.png', format='png')
```

```python
from matplotlib import pyplot
sns.set_context("notebook")
dim = (20.00, 10.00)
fig, ax = pyplot.subplots(figsize=dim)
sns.set(style="white")
# Draw a categorical scatterplot to show each observation
sns.boxplot(y="RSRP", x="DeviceManufacturer",  
              data=rsrp.groupby('DeviceManufacturer').filter(lambda x: len(x) >= 30),
              order = rsrp.groupby('DeviceManufacturer').filter(lambda x: len(x) >= 30)\
                     .groupby(['DeviceManufacturer']).median()\
                    .sort_values('RSRP', ascending = False)\
                    .iloc[0:30]
                     .index)

ax.xaxis.grid(True)
ax.yaxis.grid(False)
sns.despine(left=True, bottom=True)

#plt.ylim(0, 100000)
plt.xticks(rotation=75, fontsize=14)
ax.set_xlabel('DeviceManufacturer', fontsize=14)
ax.set_ylabel('RSRP', fontsize=14)
```
